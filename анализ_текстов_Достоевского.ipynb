```python
import re
import statistics
from collections import Counter
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
import csv
import numpy

with open('/content/Преступление и наказание.txt', 'r', encoding="utf-8") as f:
  text = f.read().lower()

clean_text = re.sub(r"[«»\"“”]", "", text)

punctuation_counts = {
    "comma": len(re.findall(r",", clean_text)),
    "dash": len(re.findall(r"—|–|-", clean_text)),
    "ellipsis": len(re.findall(r"\.\.\.", clean_text)),
    "exclamation": len(re.findall(r"!", clean_text)),
    "question": len(re.findall(r"\?", clean_text))
}

sentences = re.split(r"[.!?]+", clean_text)
sentences = [s.strip() for s in sentences if s.strip()]

word_pattern = r"[а-яёa-z]+"
word_counts_per_sentence = [len(re.findall(word_pattern, s)) for s in sentences]
total_words = sum(word_counts_per_sentence)
avg_sentence_length = statistics.mean(word_counts_per_sentence) if word_counts_per_sentence else 0

first_second_person = ["я", "мы", "ты", "вы"]
third_person = ["он", "она", "оно", "они"]

words = re.findall(word_pattern, clean_text)
word_freq = Counter(words)

pronouns_1_2 = sum(word_freq[p] for p in first_second_person)
pronouns_3 = sum(word_freq[p] for p in third_person)

results = {
    "total_words": total_words,
    "avg_sentence_length": round(avg_sentence_length, 2),
    "comma_count": punctuation_counts["comma"],
    "dash_count": punctuation_counts["dash"],
    "ellipsis_count": punctuation_counts["ellipsis"],
    "exclamation_count": punctuation_counts["exclamation"],
    "question_count": punctuation_counts["question"],
    "pronouns_1_2": pronouns_1_2,
    "pronouns_3": pronouns_3,
    "ratio_1_2_to_3": round(pronouns_1_2 / pronouns_3, 3) if pronouns_3 > 0 else None
}

for key, val in results.items():
    print(f"{key}: {val}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
data = {
    "Произведение": ["Двойник", "Белые ночи", "Преступление и наказание", "Идиот", "Бесы", "Братья Карамазовы"],
    "Год": [1846, 1848, 1866, 1869, 1872, 1880],
    "Средняя_длина_предложения": [14, 13, 14, 16, 19, 25],
    "Частота_тире": [556, 597, 420, 477, 231, 181],
    "Частота_вос.знаков": [96, 237, 162, 91, 58, 9],
    "Частота_воп.знаков": [56, 172, 90, 84, 30, 23],
    "Частота_троеточий": [124, 128, 136, 40, 17, 10],
    "Отношение м": [1, 2.9, 0.9, 0.6, 0.4, 0.2],
}

df = pd.DataFrame(data)
```

```python
print(df)
```

```python
plt.figure()
plt.plot(df["Год"], df["Средняя_длина_предложения"], marker="o")
plt.title("Средняя длина предложения")
plt.xlabel("Год")
plt.ylabel("Средняя длина предложения (в словах)")
plt.grid(False)
plt.show()
```

```python
a = df["Частота_тире"]
b = df["Частота_вос.знаков"]
d = df["Частота_воп.знаков"]
e = df["Частота_троеточий"]
c = df["Год"]

plt.figure(figsize=(10, 6))
sns.lineplot(x=c, y=a, label="--", color='black')
sns.lineplot(x=c, y=b, label="!", color='red')
sns.lineplot(x=c, y=d, label="?", color='blue')
sns.lineplot(x=c, y=e, label="...", color='green')
plt.xlabel("год")
plt.ylabel("")
plt.title("Количество экспрессивных знаков препинания")
plt.show()
```

```python
m = df["Средняя_длина_предложения"]
zh = df["Отношение м"]
c = df["Год"]

plt.figure(figsize=(10, 6))
sns.lineplot(x=c, y=m, label="ср. длина предложений", color='black')
sns.lineplot(x=c, y=zh, label="отношение местоимений", color='red')
plt.xlabel("год")
plt.ylabel("")
plt.title("связь длины предложений с местоимениями")
plt.show()
```

```python
from scipy.stats import pearsonr

r, p = pearsonr(df["Отношение м"], df["Средняя_длина_предложения"])
print(f"Коэффициент корреляции: {r:.3f}, p-value: {p:.4f}")
```

```python
r, p = pearsonr(df["Год"], df["Средняя_длина_предложения"])
print(f"Коэффициент корреляции: {r:.3f}, p-value: {p:.4f}")
```
